```python

#importing packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()

pd.set_option('display.float_format', lambda x: '%.2f' % x)
pd.set_option('display.max_columns', 21)
pd.set_option('display.max_rows', 70)

df = pd.read_csv("kc_house_data.csv") 
df.head(3)

#checking if there is any missing data
df.isnull().values.any() 
sns.pairplot(data=df, x_vars=['sqft_living','sqft_lot','sqft_above','sqft_living15','sqft_lot15'], y_vars=["price"])

df2 = df[["price", "sqft_living"]]
sns.jointplot('sqft_living','price', data=df2, size=8, alpha=.5, marker='+')
sns.jointplot('sqft_living','price', data=df2, xlim=(500,3500), ylim=(100000,1000000), size=8, alpha=.5, marker='+')
```
